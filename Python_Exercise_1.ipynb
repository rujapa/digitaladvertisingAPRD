# Question 1 When considering 'cost per impression', what day of the week works best? (What day is it cheapest to generate impressions) What day works worst?

```python
#Aggregate data to count the total for each weekday 
df.groupby('day_of_week').agg({'Cost Per Impression':['count','sum','mean','std']})
```

```python
#create model to find which variable most strongly predicts unique clicks 
X = df[['Reach','Frequency']]
y = df["Unique Clicks"]
model = sm.OLS(y, X).fit()
predictions = model.predict(X)
# Print out the statistics
print(model.summary())
```
